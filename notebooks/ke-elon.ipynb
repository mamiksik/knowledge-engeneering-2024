```python
df = pd.read_csv('/kaggle/input/elongatedmuskrat/elonmusk.csv')
```

```python
def translate_string(input_str):
    if input_str.lower() == "none":
        return []
    else:
        return input_str.lower().split(", ")
    

def extract_companies_mentioned(tweet):
    """Returns the names of any company mentioned in the tweet"""
    
    prefix = ("Are there any companies mentioned in the following tweet? If yes, list all the companies, seperated by "
              "a comma and a space (e.g. 'Apple, Tesla, Samsung'), if no, say 'None'. Please say nothing other than this. \n")
    
    response = ollama.generate(model='llama3', prompt=prefix+tweet)
    return translate_string(response['response'])
```

```python
df['Companies'] = df['Tweets'].apply(lambda x: extract_companies_mentioned(x))
df.to_csv('elon_with_companies.csv', index=False)
print("Success")
```
